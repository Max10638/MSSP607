```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams.update({
    'figure.figsize': (7.5, 7.5),
    'axes.spines.right': False,
    'axes.spines.left': False,
    'axes.spines.top': False,
    'axes.spines.bottom': False})

import random
from numpy import random as nprand
seed = hash("Network Science in Python") % 2**32
nprand.seed(seed)
random.seed(seed)
```

```python
import networkx as nx

G = nx.Graph()

G.add_node('Mingxuan Li')
G.add_nodes_from(['Naiqian Zhang', 'Qiyao Huang', 'Longsha Wei', 'Astrid Fan'])

G.add_edge('Mingxuan Li', 'Naiqian Zhang')
G.add_edges_from([('Mingxuan Li', 'Qiyao Huang'), ('Mingxuan Li', 'Longsha Wei'), ('Mingxuan Li', 'Astrid Fan'),
                  ('Naiqian Zhang', 'Qiyao Huang'), ('Naiqian Zhang', 'Longsha Wei'), ('Naiqian Zhang', 'Astrid Fan'),
                   ('Qiyao Huang', 'Longsha Wei'), ('Qiyao Huang', 'Astrid Fan'), ('Longsha Wei', 'Astrid Fan')])

plt.figure(figsize=(7.5, 7.5))
nx.draw_networkx(G)
```
